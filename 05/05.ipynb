# 函数入门

本教程将带你系统学习 Python 函数的定义、参数、返回值、作用域、嵌套与递归调用等核心知识。

## 1. 函数的概念及分类

### 1.1 函数的概念
函数是**组织好**的、可**重复使用**的、用于执行**特定任务**的代码块。

**生活类比**：就像智能家居中的“场景模式”，提前设置好一系列操作，需要时一键启动。

使用函数的主要优势：
- 减少重复代码
- 提高可读性和可维护性
- 便于调试和模块化

### 1.2 Python 中函数的分类

- **内置函数**：Python 自带的函数，如 `print()`、`len()`、`sum()` 等。
  - 官方文档：https://docs.python.org/zh-cn/3/library/functions.html
- **模块提供的函数**：需要导入模块才能使用的函数，如 `math.sqrt()`、`random.randint()` 等。
  - 官方文档：https://docs.python.org/zh-cn/3/py-modindex.html
- **自定义函数**：由用户根据需求自行定义的函数。

本章重点讲解自定义函数。

## 2. 函数的基本使用

### 2.1 定义函数

**语法**：
```python
def 函数名():
    函数体
```
- `def` 是定义函数的关键字。
- 函数名遵循标识符命名规范。
- 函数定义后不会立即执行，需要调用才会运行。

````python
# 练习 01：函数基本使用
def welcome():
    print("欢迎来到 Python 课堂！")
    print("Python 让编程更简单！")

# 调用函数
welcome()
````

### 2.2 调用函数

**语法**：`函数名()`

函数必须先定义再调用。

````python
# 多次调用
welcome()
welcome()
````

## 3. 函数的参数

参数可以让函数接收外部数据，使函数更通用、更灵活。

### 3.1 参数的作用

没有参数时，函数只能处理固定的数据：

````python
def order():
    num = 1
    dish = "辣椒炒肉"
    print(f"您点的是：{num}份 {dish}")

order()
````

有了参数，就可以灵活修改：

````python
# 练习 02：参数的使用
def order(num, dish):
    print(f"您点的是：{num}份 {dish}")
    print(f"{dish}可是很好吃的！")
    print(f"你只点了{num}份，够吃吗？\n")

order(1, "辣椒炒肉")
order(2, "辣子鸡")
````

### 3.2 形参与实参

- **形参（形式参数）**：定义函数时的变量，用于接收数据。
- **实参（实际参数）**：调用函数时传入的具体值。

形参只在函数内部有效。

### 3.3 位置参数

按照形参出现的顺序，依次传入实参。

````python
# 练习 03：位置参数
def order(num, dish):
    print(f"您点的是：{num}份 {dish}")

order(3, "宫保鸡丁")
# 下面的调用会出错或导致逻辑错误，请勿模仿：
# order(3)                # 参数个数不对
# order("宫保鸡丁", 3)    # 顺序错误
````

### 3.4 关键字参数

调用时通过 `形参名=值` 的方式传递，顺序可以任意。

````python
# 练习 04：关键字参数
def greet(name, gender, age, height):
    print(f"我叫{name}，性别{gender}，年龄{age}，身高{height}cm")

greet(name="张三", gender="男", age=18, height=172)
greet(height=172, age=18, gender="男", name="张三")
````

**位置参数和关键字参数可以混用，但位置参数必须在前。**

````python
greet("张三", "男", height=172, age=18)  # 正确
# greet(height=172, age=18, "张三", "男")  # 错误
````

### 3.5 限制传参方式

- `/` 前面的参数只能用位置参数传递。
- `*` 后面的参数只能用关键字参数传递。

````python
# 练习 05：限制传参方式
def greet(name, /, gender, *, age, height):
    print(f"我叫{name}，性别{gender}，年龄{age}，身高{height}cm")

greet("张三", "男", age=18, height=172)          # 正确
greet("张三", gender="男", age=18, height=172)  # 正确
# greet(name="张三", gender="男", age=18, height=172)  # 错误
````

### 3.6 参数默认值

定义函数时可以为形参指定默认值，调用时若不传则使用默认值。
**注意**：默认参数必须放在必选参数后面。

````python
# 练习 06：参数默认值
def greet(name, gender, age, height, msg="你好"):
    print(f"我叫{name}，性别{gender}，年龄{age}，身高{height}cm")
    print(f"我想说：{msg}")

greet("张三", "男", 18, 172)           # 使用默认 msg
greet("张三", "男", 18, 172, "hello") # 覆盖默认值
````

### 3.7 可变参数

- `*args` 接收任意数量的**位置参数**，打包为**元组**。
- `**kwargs` 接收任意数量的**关键字参数**，打包为**字典**。

````python
# 练习 07：可变参数
def test1(*args):
    print("位置参数元组:", args)

test1("张三", "男", 18, 172)

def test2(**kwargs):
    print("关键字参数字典:", kwargs)

test2(name="张三", gender="男", age=18, height=172)

# 同时使用：*args 必须在 **kwargs 前面
def test3(a, b, *args, c="Python", **kwargs):
    print("a:", a)
    print("b:", b)
    print("c:", c)
    print("args:", args)
    print("kwargs:", kwargs)

test3("张三", "男", "抽烟", "喝酒", age=18, height=172)
````

### 3.8 特殊的字面量 None

`None` 表示空值、无值。类型是 `NoneType`。
- 在布尔判断中视为 `False`。
- 不能参与数学运算或字符串拼接。
- 函数默认返回 `None`。

````python
# 练习 08：None
msg = None
print("类型:", type(msg))
print("布尔值:", bool(msg))

if not msg:
    print("msg 为空")

# 无法这样操作：
# result = msg + 1
# result = msg + "hello"
````

## 4. 返回值

函数执行完后可以将结果返回给调用者。
使用 `return` 关键字：
- 结束函数运行。
- 将 `return` 后面的值作为函数的返回值。
如果不写 `return` 或者只写 `return`，函数返回 `None`。

````python
# 练习 09：返回值
def add(n1, n2):
    print(f"计算 {n1} + {n2}")
    return n1 + n2

result = add(100, 200)
print("结果:", result)

# 没有返回值的函数
res = print("hello")
print("print 的返回值:", res)
````

## 5. 全局作用域与局部作用域

- **全局变量**：在函数外部定义的变量，整个程序都可访问。
- **局部变量**：在函数内部定义的变量，仅在函数内有效。
当函数内部需要修改全局变量时，使用 `global` 关键字声明。

````python
# 练习 10：全局与局部作用域
a = 100
b = 200

def test():
    c = "Python"
    d = "你好"
    global a
    a = 300
    print("函数内 a:", a)
    print("函数内 b:", b)
    print("函数内 c:", c)
    print("函数内 d:", d)

test()
print("*" * 20)
print("全局 a:", a)
print("全局 b:", b)
# print(c)  # 错误，局部变量无法在全局访问
````

**局部变量的生命周期**：函数调用时创建，执行结束后销毁。
**全局变量**：程序开始时创建，程序结束后销毁。

````python
# 演示生命周期
def test2():
    m = 100
    m += 1
    print(f"test2 中的 m: {m}")

test2()
test2()
test2()  # 每次都重新初始化

n = 100
def test3():
    global n
    n += 1
    print(f"test3 中的 n: {n}")

test3()
test3()
test3()
print("最终的全局 n:", n)
````

## 6. 函数嵌套调用

一个函数在执行过程中调用另一个函数，称为嵌套调用。

````python
# 练习 11：嵌套调用
def greet(name, msg):
    print(f"我叫{name}，我想说的话在下面：")
    speak(msg)
    print("我想说的结束了")

def speak(msg):
    print("---" * 10)
    print(msg)
    print("---" * 10)

greet("张三", "你好啊")

# 更深的嵌套
def test1():
    print("进入 test1")
    test2()
    print("退出 test1")

def test2():
    print("进入 test2")
    test3()
    print("退出 test2")

def test3():
    print("进入 test3")
    print("*** 正在执行 test3 ***")
    print("退出 test3")

test1()
````

## 7. 递归调用

递归是指函数自己调用自己。**必须有终止条件**，否则会无限循环。

````python
# 练习 12：递归调用（有终止条件）
def welcome(n):
    print(f"你好啊 {n}")
    if n > 1:
        welcome(n - 1)

welcome(5)

# 从小到大打印
def welcome_asc(n):
    if n > 1:
        welcome_asc(n - 1)
    print(f"你好啊 {n}")

welcome_asc(5)
````

### 递归的应用：求阶乘

````python
# 练习 13：递归求阶乘
def factorial(num):
    if num == 0:
        return 1
    else:
        return num * factorial(num - 1)

print("5! =", factorial(5))
print("6! =", factorial(6))
````

## 8. 函数说明文档

在函数内部使用三引号字符串编写文档，描述功能、参数和返回值。

````python
# 练习 14：函数说明文档
def add(n1, n2):
    """
    计算两个数相加的结果。
    :param n1: 第一个数
    :param n2: 第二个数
    :return: 二者相加的结果
    """
    return n1 + n2

result = add(1, 2)
print(result)
# 在 Jupyter 中可以按 Shift+Tab 查看文档
help(add)
````

## 9. 综合案例：健身挑战赛

结合参数、返回值、可变参数、默认参数等知识，完成一个简易的健身挑战统计程序。

````python
# 练习 15：综合案例
def calc_total(*nums):
    """
    计算总运动量
    :param nums: 每一天的运动量（可变参数）
    :return: 总运动量
    """
    return sum(nums)

def calc_avg(total, days=7):
    """
    计算平均值
    :param total: 总运动量
    :param days: 天数（默认7）
    :return: 平均值
    """
    return total / days

def check_success(total, goal=120):
    """
    判断挑战是否成功
    :param total: 总运动量
    :param goal: 成功阈值（默认120）
    :return: 结果字符串
    """
    if total >= goal:
        return "✅ 恭喜！挑战成功！"
    else:
        return "❌ 抱歉！挑战失败！"

def main(title, duration, goal):
    """
    主函数：开始一场挑战赛
    :param title: 比赛标题
    :param duration: 持续天数
    :param goal: 目标运动量
    """
    print(f"【{title}】【{duration}天】挑战赛（请输入每天的数量）")
    num1 = int(input("第1天："))
    num2 = int(input("第2天："))
    num3 = int(input("第3天："))
    
    total = calc_total(num1, num2, num3)
    avg = calc_avg(total, duration)
    result = check_success(total, goal)
    
    print(f"\n【{title}】【{duration}天】健身总结")
    print(f"总数：{total}，平均值：{avg:.1f}")
    print(result)

# 启动挑战
main("俯卧撑", 3, 40)
````

### 01 字面量（Literal）
字面量是直接写在代码中的固定值，比如数字 `100`、字符串 `"hello"`、布尔值 `True` 等。Python 解释器会原样识别它们。


```python
# 数字字面量
100
3.14

# 字符串字面量
"Hello Python"
'你好'

# 布尔字面量
True
False

# 列表字面量
[1, 2, 3]
```

### 02 变量（Variable）

变量是用来存储数据的容器。在 Python 中，你不需要声明类型，直接赋值即可创建变量。

```python
# 定义变量
name = "Alice"
age = 25
height = 1.68

# 使用变量
print(name)
print(age)
print(height)

# 修改变量的值
age = 26
print(age)
```

### 03 标识符命名规则

标识符就是我们给变量、函数、类等起的名字，需要遵守以下规则：
- 只能由字母、数字、下划线 `_` 组成
- 不能以数字开头
- 不能使用 Python 关键字（如 `if`、`for`、`while`）
- 区分大小写（`Name` 和 `name` 是不同的）

```python
# 合法命名
user_name = "Tom"
_score = 98
var1 = 100

# 非法命名（以下代码会报错，请勿直接运行）
# 1st_place = "gold"   # 不能以数字开头
# my-name = "abc"      # 不能包含连字符
# if = 10              # 不能使用关键字
```

> **命名习惯**：变量名建议使用小写字母加下划线（snake_case），类名使用大写字母开头的驼峰命名法（PascalCase）。


### 04 常量（Constant）

Python 没有真正意义上的常量，通常用全大写字母的变量名来表示“约定俗成”的常量，提示别人不要修改它。


```python
PI = 3.14159
MAX_SPEED = 120
GRAVITY = 9.8

print(PI)
```

### 05 注释（Comment）

注释是对代码的解释说明，不会被解释器执行。单行注释用 `#`，多行注释可以用三个单引号或双引号。


```python
# 这是单行注释
print("Hello")  # 行尾注释也可以

"""
这是多行注释（其实是一个未被赋值的多行字符串）
可以写很多行
"""
print("注释示例")
```

### 06 字符编码

Python 3 默认使用 UTF-8 编码，字符串在内存中以 Unicode 形式存储，因此可以直接处理中文、emoji 等各种字符。


```python
# 查看字符的 Unicode 编码值
print(ord("中"))   # 输出 20013
print(ord("A"))    # 输出 65

# 将编码值转换为字符
print(chr(20013))  # 输出 '中'
```

### 07 数据类型（Data Type）

Python 中常见的数据类型：

| 类型 | 示例 | 说明 |
|------|------|------|
| int | 10, -5, 0 | 整数 |
| float | 3.14, -0.5 | 浮点数（小数） |
| str | "hello", 'abc' | 字符串 |
| bool | True, False | 布尔值 |
| list | [1, 2, 3] | 列表 |
| tuple | (1, 2, 3) | 元组 |
| dict | {"name": "Tom"} | 字典 |
| set | {1, 2, 3} | 集合 |

```python
# 使用 type() 查看数据类型
a = 10
b = 3.14
c = "Python"
d = True

print(type(a))
print(type(b))
print(type(c))
print(type(d))
```

### 08 整型（int）

整型就是整数，Python 中的整数没有大小限制（只受内存限制），可以表示任意大的整数。

```python
# 整数定义
num1 = 100
num2 = -50
big_num = 12345678901234567890

print(num1, num2, big_num)
print(type(big_num))
```

### 09 浮点型（float）

浮点数即带有小数点的数字，Python 中采用双精度浮点数（IEEE 754），运算时可能存在微小的精度误差。


```python
# 浮点数定义
pi = 3.14159
negative = -0.003
sci = 1.2e5    # 科学计数法，表示 1.2 × 10⁵ = 120000.0

print(pi)
print(sci)

# 浮点数精度问题
print(0.1 + 0.2)   # 输出 0.30000000000000004
```

### 10 字符串的四种定义方式

字符串可以使用单引号、双引号、三单引号或三双引号定义。多行字符串通常使用三引号。


```python
# 四种方式
s1 = '单引号字符串'
s2 = "双引号字符串"
s3 = '''三单引号字符串
可以换行'''
s4 = """三双引号字符串
也可以换行"""

print(s1)
print(s2)
print(s3)
print(s4)

# 引号嵌套的技巧
s5 = "I'm a student"          # 双引号内可以直接写单引号
s6 = 'He said: "Hello"'       # 单引号内可以直接写双引号
print(s5, s6)
```

### 11 字符串格式化输出

Python 提供多种字符串格式化方式，最推荐使用 **f-string**（Python 3.6+）。

```python
name = "Alice"
age = 25

# 1. % 格式化（旧式）
print("我叫 %s，今年 %d 岁" % (name, age))

# 2. str.format() 方法
print("我叫 {}，今年 {} 岁".format(name, age))

# 3. f-string（推荐）
print(f"我叫 {name}，今年 {age} 岁")
```

### 12 字符串占位符精度控制

在格式化时，可以通过格式说明符控制数字的精度、宽度、对齐等。


```python
pi = 3.1415926

# 保留两位小数
print(f"{pi:.2f}")

# 保留三位小数，总宽度为8，右对齐
print(f"{pi:8.3f}")

# 左对齐
print(f"{pi:<8.3f}")

# 百分比显示
rate = 0.856
print(f"合格率：{rate:.1%}")   # 输出 85.6%
```

### 13 字符串转义字符

转义字符以反斜杠 `\` 开头，用于表示一些特殊字符。

| 转义字符 | 含义 |
|----------|------|
| `\n` | 换行 |
| `\t` | 制表符（Tab） |
| `\\` | 反斜杠本身 |
| `\'` | 单引号 |
| `\"` | 双引号 |

```python
print("Hello\nWorld")
print("Name:\tAlice")
print("路径：C:\\Users\\Alice")
print("他说：\"你好\"")
```

### 14 数据类型转换

可以使用 `int()`、`float()`、`str()`、`bool()` 等函数进行类型转换。

```python
# 字符串转数字
num_str = "123"
num = int(num_str)
print(num, type(num))

# 数字转字符串
age = 25
age_str = str(age)
print(age_str, type(age_str))

# 浮点数转整数（直接截断小数部分）
pi = 3.9
print(int(pi))   # 输出 3

# 布尔转换：非0为True，0为False；空字符串为False，非空为True
print(bool(1))   # True
print(bool(0))   # False
print(bool(""))  # False
```

### 15 算术运算符

| 运算符 | 描述 | 示例 |
|--------|------|------|
| `+` | 加 | `1 + 2` → `3` |
| `-` | 减 | `5 - 2` → `3` |
| `*` | 乘 | `3 * 4` → `12` |
| `/` | 除（结果为浮点数） | `10 / 2` → `5.0` |
| `//` | 整除（向下取整） | `10 // 3` → `3` |
| `%` | 取余 | `10 % 3` → `1` |
| `**` | 幂运算 | `2 ** 3` → `8` |

```python
a = 10
b = 3

print("加法:", a + b)
print("除法:", a / b)
print("整除:", a // b)
print("取余:", a % b)
print("幂:", a ** b)
```

### 16 赋值运算符

最基本的赋值运算符是 `=`，它将右侧的值赋给左侧的变量。

```python
x = 10
y = x
print(x, y)

# 同时给多个变量赋值
a, b, c = 1, 2, 3
print(a, b, c)

# 交换两个变量的值
a, b = b, a
print(a, b)
```

### 17 复合赋值运算符

复合赋值运算符将算术运算和赋值结合在一起，使代码更简洁。

| 运算符 | 等价于 |
|--------|--------|
| `+=` | `x = x + 值` |
| `-=` | `x = x - 值` |
| `*=` | `x = x * 值` |
| `/=` | `x = x / 值` |
| `//=` | `x = x // 值` |
| `%=` | `x = x % 值` |
| `**=` | `x = x ** 值` |

```python
num = 10
num += 5   # 相当于 num = num + 5
print(num)

num *= 2
print(num)

num //= 3
print(num)
```

### 18 比较运算符

比较运算符用于比较两个值，结果返回布尔值 `True` 或 `False`。

| 运算符 | 描述 |
|--------|------|
| `==` | 等于 |
| `!=` | 不等于 |
| `>` | 大于 |
| `<` | 小于 |
| `>=` | 大于等于 |
| `<=` | 小于等于 |

```python
a = 10
b = 20

print(a == b)   # False
print(a != b)   # True
print(a < b)    # True
print(a >= 10)  # True
```

### 19 布尔类型（bool）

布尔类型只有两个值：`True` 和 `False`。它们是 Python 的关键字，首字母必须大写。

```python
is_student = True
is_raining = False

print(is_student)
print(type(is_student))

# 比较运算的结果是布尔值
result = 5 > 3
print(result)
```

### 20 逻辑运算符

逻辑运算符用于组合多个布尔表达式。

| 运算符 | 描述 | 示例 |
|--------|------|------|
| `and` | 与（两边都为 True 才为 True） | `True and False` → `False` |
| `or` | 或（至少一边为 True 则为 True） | `True or False` → `True` |
| `not` | 非（取反） | `not True` → `False` |

```python
age = 20
has_ticket = True

# 判断能否入场：年龄>=18 且 有票
can_enter = age >= 18 and has_ticket
print(can_enter)

# 判断是否免费：年龄<12 或 年龄>=60
is_free = age < 12 or age >= 60
print(is_free)

print(not is_free)
```

### 21 进制

Python 中可以用不同前缀表示不同进制的整数：
- 二进制：`0b` 开头
- 八进制：`0o` 开头
- 十六进制：`0x` 开头

```python
# 不同进制的字面量
bin_num = 0b1010   # 二进制 1010，十进制为 10
oct_num = 0o12     # 八进制 12，十进制为 10
hex_num = 0xA      # 十六进制 A，十进制为 10

print(bin_num, oct_num, hex_num)

# 转换函数
x = 42
print(bin(x))   # 转为二进制字符串
print(oct(x))   # 转为八进制字符串
print(hex(x))   # 转为十六进制字符串
```

### 22 输入语句（input）

`input()` 函数用于接收用户从键盘输入的内容，返回的始终是字符串类型。

```python
# 基本用法
name = input("请输入你的名字：")
print(f"你好，{name}！")

# 获取数字输入（需要类型转换）
age_str = input("请输入你的年龄：")
age = int(age_str)
print(f"明年你就 {age + 1} 岁了。")
```

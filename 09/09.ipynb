## 01 错误与异常

### 1.1 错误与异常的区别

- **错误**：通常是代码本身存在语法问题，解释器无法正常解析。
- **异常**：代码语法没有问题，但程序在运行时因为数据、对象、索引、变量等原因发生问题。

```python
show_title("1. 错误与异常")

print("【语法错误示例：仅展示，不执行】")
bad_code = '''
age = 18
if age >= 18
    print("成年人")
'''
print(bad_code)
print("说明：上述代码缺少冒号，属于语法错误。语法错误需要修改代码本身，不能依赖普通异常处理机制解决。")
```

## 02 异常处理

异常处理不是让异常消失，而是将异常捕获起来，并根据不同情况执行合适的处理逻辑，避免程序直接崩溃。

```python
show_title("2. 异常处理")
show_subtitle("2.1 为什么要进行异常处理")

def program_without_handler(a_text, b_text):
    print("欢迎使用本程序")
    a = int(a_text)
    b = int(b_text)
    result = a / b
    print(f"{a} 除以 {b} 的结果是：{result}")
    print("*******我是后续的其它逻辑1*******")
    print("*******我是后续的其它逻辑2*******")

for a_text, b_text in [("10", "2"), ("10", "0"), ("abc", "2")]:
    print(f"\n测试数据：a={a_text}, b={b_text}")
    try:
        program_without_handler(a_text, b_text)
    except Exception as e:
        print(f"程序中断，异常类型：{type(e).__name__}，异常信息：{e}")
```

```python
show_subtitle("2.2 异常处理初级写法")

def basic_handler(a_text, b_text):
    print(f"\n输入：a={a_text}, b={b_text}")
    try:
        a = int(a_text)
        b = int(b_text)
        result = a / b
        print(f"{a} 除以 {b} 的结果是：{result}")
    except:
        print("抱歉，程序出现了异常！")
    print("*******我是后续的其它逻辑1*******")
    print("*******我是后续的其它逻辑2*******")

basic_handler("10", "2")
basic_handler("10", "0")
basic_handler("abc", "2")
```

## 03 手动抛出异常

当程序遇到不符合业务规则的数据时，可以通过 `raise` 主动抛出异常，阻止错误数据继续传播。

```python
show_title("3. 手动抛出异常")

def check_age(age):
    if not isinstance(age, int):
        raise TypeError("年龄必须是整数")
    if age < 0 or age > 120:
        raise ValueError("年龄应该在 0~120 之间")
    return "成年" if age >= 18 else "未成年"

for age in [20, 16, -1, 130, '18']:
    print(f"\n测试年龄：{age}")
    try:
        print("判断结果：", check_age(age))
    except Exception as e:
        print(f"程序异常：{type(e).__name__} -> {e}")
```

```python
show_subtitle("补充：用户名校验示例")

def check_username(username):
    if not isinstance(username, str):
        raise TypeError("用户名必须是字符串")
    if len(username) < 3 or len(username) > 12:
        raise ValueError("用户名长度应在 3~12 个字符之间")
    if not username.isalnum():
        raise ValueError("用户名只能由字母和数字组成")
    return f"用户名“{username}”合法"

for name in ['tom123', 'ab', 'verylongusername123', '张三@', 1001]:
    print(f"\n测试用户名：{name}")
    try:
        print(check_username(name))
    except Exception as e:
        print(f"程序异常：{type(e).__name__} -> {e}")
```

## 04 异常传递机制

如果异常没有在当前函数中被捕获，它会沿着函数调用链逐层向上传递，直到被某一层处理；如果始终无人处理，程序就会终止。

```python
show_title("4. 异常的传递机制")

def test1():
    print("****** test1 开始 ******")
    result = '100' + 100
    print("****** test1 结束 ******")

def test2():
    print("****** test2 开始 ******")
    test1()
    print("****** test2 结束 ******")

def test3():
    print("****** test3 开始 ******")
    try:
        test2()
    except Exception as e:
        print(f"test3 捕获到异常：{e}")
    print("****** test3 结束 ******")

test3()
```

```python
show_subtitle("补充：中间层记录日志后继续向上抛出")

def read_config():
    print("read_config：开始读取配置文件")
    raise FileNotFoundError("config.ini 不存在")

def start_service():
    print("start_service：准备启动服务")
    try:
        read_config()
    except FileNotFoundError as e:
        print(f"start_service：记录异常信息 -> {e}")
        raise

def main():
    print("main：程序开始")
    try:
        start_service()
    except Exception as e:
        print(f"main：统一处理异常 -> {type(e).__name__} -> {e}")
    print("main：程序结束")

main()
```

## 05 自定义异常类

当内置异常类无法准确表达业务问题时，可以自定义异常类，使错误信息更清晰、更有针对性。

```python
show_title("5. 自定义异常类")

class DataValidationError(Exception):
    pass

class SchoolNameError(DataValidationError):
    def __init__(self, msg):
        super().__init__("【学校名称异常】" + msg)

def check_school_name(name):
    if not isinstance(name, str):
        raise SchoolNameError("学校名称必须是字符串")
    name = name.strip()
    if not name:
        raise SchoolNameError("学校名称不能为空")
    if len(name) < 2 or len(name) > 20:
        raise SchoolNameError("学校名称长度应在 2~20 个字符之间")
    if any(ch in name for ch in '@#￥%'):
        raise SchoolNameError("学校名称不能包含特殊符号")
    return f"学校名称“{name}”合法"

print(issubclass(SchoolNameError, DataValidationError))
print(issubclass(SchoolNameError, Exception))

for item in ['清风学院', '', 'A', '未来科技大学@@', 123]:
    print(f"\n测试学校名称：{item}")
    try:
        print(check_school_name(item))
    except SchoolNameError as e:
        print(f"程序异常：{e}")
```

```python
show_subtitle("补充：自定义密码格式异常")

class PasswordFormatError(DataValidationError):
    def __init__(self, msg):
        super().__init__("【密码格式异常】" + msg)

def check_password(password):
    if not isinstance(password, str):
        raise PasswordFormatError("密码必须是字符串")
    if len(password) < 6:
        raise PasswordFormatError("密码长度不能少于 6 位")
    if password.isdigit():
        raise PasswordFormatError("密码不能是纯数字")
    return "密码格式合法"

for pwd in ['abc123', '123456', 'abc', 888888]:
    print(f"\n测试密码：{pwd}")
    try:
        print(check_password(pwd))
    except DataValidationError as e:
        print(f"程序异常：{e}")
```

## 本章小结

1. 错误通常指语法层面的代码问题；
2. 异常是程序运行期间出现的问题；
3. `try-except` 可以捕获异常，避免程序直接崩溃；
4. `else` 适合处理成功逻辑，`finally` 适合处理收尾逻辑；
5. `raise` 可以手动抛出异常；
6. 异常会沿函数调用链向上传递；
7. 自定义异常类有助于表达更具体的业务问题。

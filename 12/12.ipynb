## 1. 文件的分类

### 1.1 纯文本文件

纯文本文件在读取和存储时，需要遵循某种**字符编码**规范，例如 UTF-8、GBK 等。虽然底层最终仍然是二进制存储，但它表达的是“可直接阅读的文本信息”。

常见的纯文本文件：

- `.txt`
- `.py`
- `.md`
- `.html`
- `.csv`

### 1.2 二进制文件

二进制文件不以“字符编码”为核心，而是遵循某种**文件格式规范**来组织内容。它们通常需要由能够识别其格式的软件进行解析后才能正确展示。

常见的二进制文件：

- `.mp3`
- `.mp4`
- `.jpg`
- `.png`
- `.doc`
- `.ppt`
- `.pdf`

### 1.3 小结

- 文本文件：关注**编码与解码**
- 二进制文件：关注**格式与字节数据**

````python
text_examples = ['note.txt', 'main.py', 'readme.md', 'page.html']
binary_examples = ['music.mp3', 'video.mp4', 'photo.jpg', 'slides.ppt']

print('纯文本文件示例：')
for name in text_examples:
    print(' -', name)

print('\n二进制文件示例：')
for name in binary_examples:
    print(' -', name)
````

## 2. 绝对路径 vs 相对路径

### 2.1 绝对路径

绝对路径从文件系统的**根位置**开始，完整描述文件或目录所在的位置。

示例：

- Windows：`D:/demo/test/a.txt`
- macOS / Linux：`/home/user/demo/a.txt`

### 2.2 相对路径

相对路径以当前**工作目录**为参照，描述目标文件相对于当前目录的位置。

示例：

- `a.txt`
- `data/info.txt`
- `../config.json`
- `../../logs/app.log`

其中：

- `.` 表示当前目录
- `..` 表示上一级目录

### 2.3 学习建议

在编写脚本时，优先理解“当前工作目录”是什么，否则容易出现“文件找不到”的问题。

````python
import os

print('当前工作目录：', os.getcwd())
print('相对路径 data/sample.txt 的绝对路径：')
print(os.path.abspath('data/sample.txt'))
````

## 3. Python 中文件操作的标准流程

文件操作通常遵循以下三步：

1. 创建文件对象
2. 操作文件（读取、写入等）
3. 关闭文件

核心函数是 `open()`。

### 3.1 `open()` 常用参数

- `file`：文件路径
- `mode`：打开模式
- `encoding`：字符编码（主要用于文本文件）

### 3.2 常见模式

- `r`：读取，默认模式
- `w`：写入，先清空原文件
- `x`：排它性创建，文件已存在则报错
- `a`：追加写入
- `b`：二进制模式
- `t`：文本模式，默认值
- `+`：更新模式，表示可读可写

例如：

- `rt`：文本读取
- `rb`：二进制读取
- `wt`：文本写入
- `a+`：追加并支持读取

````python
# 完整写法
# file = open(file='a.txt', mode='rt', encoding='utf-8')

# 简写
# file = open('a.txt', 'rt', encoding='utf-8')

# 二进制文件示例
# file = open('image.jpg', 'rb')

print('open() 是文件操作的核心入口。')
````

## 4. 读取文件

常见读取方式有：

1. `read()`
2. `readline()`
3. `for` 循环遍历文件对象
4. `readlines()`

不同方法的重点区别在于：

- 是一次性读取，还是分批读取
- 是按字符/字节读取，还是按行读取
- 是否适合大文件

### 4.1 `read()` 方法

`read(size)` 用于读取文件内容。

- 不传 `size`：读取全部内容
- 传入 `size`：读取指定数量的字符或字节
- 多次调用会从上一次结束的位置继续读
- 读到文件末尾后，再读会得到空字符串 `''` 或空字节串 `b''`

对于大文件，不建议一次性全部读入内存。更推荐循环分批读取。

````python
# 示例：分步读取文本文件
# 前提：当前目录下存在 demo_read.txt

try:
    file = open('demo_read.txt', 'rt', encoding='utf-8')
    r1 = file.read(2)
    r2 = file.read(3)
    r3 = file.read(4)
    r4 = file.read()

    print('第一次读取：', r1)
    print('第二次读取：', r2)
    print('第三次读取：', r3)
    print('剩余内容：', r4)
    file.close()
except FileNotFoundError:
    print('未找到 demo_read.txt，请先创建该文件后再运行。')

print('\n适合大文件的分批读取示例：')
print('''
with open('demo_read.txt', 'rt', encoding='utf-8') as file:
    while True:
        result = file.read(10)
        if result == '':
            break
        print(result, end='')
''')
````

### 4.2 `readline()` 方法

`readline(size)` 用于读取**一行**内容。

- 不传参数：读取当前整行
- 传参数：读取当前行中最多指定数量的字符或字节
- 参数不是“行数”
- 读到文件末尾后，再读返回空字符串

它适合需要“一行一行”处理文本的场景。

````python
try:
    file = open('demo_readline.txt', 'rt', encoding='utf-8')
    r1 = file.readline()
    r2 = file.readline()
    r3 = file.readline()
    r4 = file.readline()

    print(r1.strip())
    print(r2.strip())
    print(r3.strip())
    print(r4.strip())
    file.close()
except FileNotFoundError:
    print('未找到 demo_readline.txt，请先准备测试文件。')

print('\n循环读取每一行：')
print('''
with open('demo_readline.txt', 'rt', encoding='utf-8') as file:
    while True:
        line = file.readline()
        if line == '':
            break
        print(line, end='')
''')
````

### 4.3 `for` 循环遍历文件对象

Python 中文件对象本身就是可迭代对象，因此可以直接用 `for` 循环逐行遍历。

这是非常常见、也非常推荐的读取方式之一，因为它语义清晰，写法简洁。

````python
try:
    file = open('demo_for.txt', 'rt', encoding='utf-8')
    for line in file:
        print(line, end='')
    file.close()
except FileNotFoundError:
    print('未找到 demo_for.txt。')
````

### 4.4 `readlines()` 方法

`readlines(hint)` 会一次性读取多行内容，并返回一个列表。

- 不传 `hint`：读取全部行
- 传 `hint`：按期望字符数或字节数读取一部分内容
- 返回值是列表，列表中的每个元素通常对应文件中的一行

注意：由于它通常会把内容一次性读入内存，因此**不适合大文件**。

````python
try:
    file = open('demo_lines.txt', 'rt', encoding='utf-8')
    result = file.readlines()
    print(result)
    file.close()
except FileNotFoundError:
    print('未找到 demo_lines.txt。')
````

### 4.5 最佳实践

更推荐使用 `with` 上下文管理器配合 `for` 循环逐行读取文件：

- 代码更简洁
- 能自动关闭文件
- 对大多数文本处理场景都很友好

````python
print('推荐写法：')
print('''
with open('a.txt', 'rt', encoding='utf-8') as file:
    for line in file:
        print(line, end='')
''')
````

## 5. 关于 `with`

`with` 主要用于管理程序中那些“需要成对出现”的操作，例如：

- 打开 / 关闭
- 加锁 / 解锁
- 借用 / 归还

`with` 的价值在于：开发者只需要专注于“具体业务逻辑”，进入与退出时的清理工作交给 Python 自动完成。

语法格式：

```python
with 表达式 as 变量:
    具体操作
```

### 上下文管理器协议

- `__enter__()`：进入 `with` 代码块之前调用
- `__exit__()`：离开 `with` 代码块之后调用，即使发生异常也会调用

这使得 `with` 非常适合资源管理。

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def speak(self):
        print(f'我叫{self.name}，年龄是{self.age}')

    def __enter__(self):
        print('-----进入逻辑-----')
        return self

    def __exit__(self, exc_type, exc_val, exc_tb):
        print('-----离开逻辑-----')
        if exc_type:
            print(f'异常类型：{exc_type}')
            print(f'异常对象：{exc_val}')
            print(f'异常追踪信息：{exc_tb}')
        return True

with Person('张三', 18) as p1:
    p1.speak()
    print(666)
````

## 6. 写入文件

常见写入模式有 `w`、`x`、`a`。

### 6.1 `w` 模式

- 用于写入
- 如果文件已存在，会先清空原内容
- 如果文件不存在，通常会创建新文件

### 6.2 `x` 模式

- 排它性创建
- 文件已存在则直接报错
- 适合不希望覆盖原文件的场景

### 6.3 `a` 模式

- 追加写入
- 如果文件存在，则在末尾继续写
- 如果文件不存在，通常会创建新文件

````python
# w 模式：写入前清空文件
with open('b.txt', 'wt', encoding='utf-8') as file:
    file.write('你好，w 模式会覆盖原内容')

# x 模式：文件存在则报错
# with open('demo.txt', 'xt', encoding='utf-8') as file:
#     file.write('你好，x 模式要求文件必须不存在')

# a 模式：末尾追加
with open('a.txt', 'at', encoding='utf-8') as file:
    file.write('\n你好，a 模式会在末尾追加')

print('写入示例执行完毕。')
````

## 7. `flush()` 方法

写文件时，数据通常不会每写一次就立刻保存到磁盘，而是先进入**缓冲区**。

`flush()` 的作用是：

- 立即把缓冲区中的内容写入文件
- 适合日志、长时间运行任务、实时写入场景

注意：`flush()` 是“尽快写入”，但并不等于所有底层缓存机制都彻底完成；不过在日常开发中，它已经非常实用。

````python
import time

with open('demo_flush.txt', 'at', encoding='utf-8') as file:
    file.write('你好1\n')
    file.write('你好2\n')
    file.flush()
    print('前两行已经主动 flush 到文件。')
    # time.sleep(5)  # 演示时可适当缩短等待时间
    file.write('你好3\n')
    file.write('你好4\n')

print('flush 示例结束。')
````

## 8. 组合模式

`+` 表示“更新模式”，即文件既可以读，也可以写。

常见组合：

- `r+`：可读可写，文件必须已存在
- `w+`：可读可写，但会先清空文件
- `x+`：可读可写，要求文件必须不存在
- `a+`：可读可写，写入时默认追加到末尾

### 关于 `seek()`

`seek(offset, whence)` 用于移动文件指针。

- `offset`：偏移量
- `whence=0`：从开头计算
- `whence=1`：从当前位置计算
- `whence=2`：从末尾计算

在**文本模式**下，不要随意对包含中文的文本按字符字节位置做复杂定位，否则可能造成理解偏差。

````python
# r+ 模式示例：要求文件已存在
print('r+ 示例：')
print('''
with open('a.txt', 'rt+', encoding='utf-8') as file:
    file.seek(0, 0)
    file.write('你好')
''')

# w+ 模式示例
with open('demo_w_plus.txt', 'wt+', encoding='utf-8') as file:
    file.write('你好，w+ 模式')
    file.seek(0, 0)
    result = file.read()
    print('w+ 读到的内容：', result)

print('\nx+ 示例：')
print('''
with open('demo_x_plus.txt', 'xt+', encoding='utf-8') as file:
    file.write('你好')
    file.seek(0, 0)
    result = file.read()
    print(result)
''')

with open('demo_a_plus.txt', 'at+', encoding='utf-8') as file:
    file.write('追加一行内容\n')
    file.seek(0, 0)
    result = file.read()
    print('a+ 读到的内容：')
    print(result)
````

## 9. 目录操作

目录操作通常借助 `os` 和 `shutil` 模块。

### 常见函数

- `os.mkdir(path)`：创建单级目录
- `os.makedirs(path)`：创建多级目录
- `os.rmdir(path)`：删除空目录
- `os.removedirs(path)`：递归删除空目录
- `os.path.exists(path)`：判断路径是否存在
- `os.path.isdir(path)`：判断是否为目录
- `os.path.isfile(path)`：判断是否为文件
- `os.scandir(path)`：扫描目录内容
- `os.walk(path)`：递归遍历目录树
- `shutil.rmtree(path)`：删除非空目录（危险操作）

### 使用提醒

执行删除类操作前，一定要确认路径正确，避免误删重要数据。

````python
import os
import shutil

base_dir = 'demo_dir'
sub_dir = os.path.join(base_dir, 'aa', 'bb')

# 创建多级目录
if not os.path.exists(sub_dir):
    os.makedirs(sub_dir)
    print('已创建目录：', sub_dir)

print('路径是否存在：', os.path.exists(sub_dir))
print('是否为目录：', os.path.isdir(sub_dir))
print('是否为文件：', os.path.isfile(sub_dir))

print('\n扫描 base_dir：')
for item in os.scandir(base_dir):
    print('目录' if item.is_dir() else '文件', item.name)

print('\nos.walk 递归遍历：')
for root, dirs, files in os.walk(base_dir):
    print('当前目录：', root)
    print('子目录：', dirs)
    print('文件：', files)
    print('-' * 30)

# 演示结束后可按需删除
# shutil.rmtree(base_dir)  # 危险操作，取消注释前请确认
````

## 10. 小练习 1：复制二进制文件

需求：将一个二进制文件复制到指定位置。

关键点：

- 源文件用 `rb` 打开
- 目标文件用 `wb` 打开
- 使用循环分批读取，避免一次性加载大文件
- 目标目录不存在时，先创建目录

````python
import os

source = 'music.mp3'
target = 'media_output'

if not os.path.isdir(target):
    os.makedirs(target)

target_file = os.path.join(target, 'my_music.mp3')

try:
    with open(source, 'rb') as f1, open(target_file, 'wb') as f2:
        while True:
            data = f1.read(1024)
            if not data:
                break
            f2.write(data)
    print('复制完毕：', target_file)
except FileNotFoundError:
    print('未找到源文件 music.mp3，请先准备测试文件。')
````

## 11. 小练习 2：登录日志记录

需求：

- 用户名不存在，提示“用户名未注册”，并记录日志
- 用户名存在但密码错误，提示“密码错误”，并记录日志
- 用户名和密码都正确，提示“登录成功”，并记录日志

这个练习体现了文件追加写入在实际业务中的常见用途：**日志记录**。

````python
import time

users = {
    '张三': '123456',
    '李四': '888888',
    '王五': 'abc123'
}

username = input('请输入用户名：')
password = input('请输入密码：')
now = time.strftime('%Y-%m-%d %H:%M:%S')

if username not in users:
    print('用户名未注册')
    with open('log.txt', 'at', encoding='utf-8') as file:
        file.write(f'{now}  {username}  登录失败（用户未注册）\n')
elif users[username] != password:
    print('密码错误')
    with open('log.txt', 'at', encoding='utf-8') as file:
        file.write(f'{now}  {username}  登录失败（密码错误）\n')
else:
    print('登录成功！')
    with open('log.txt', 'at', encoding='utf-8') as file:
        file.write(f'{now}  {username}  登录成功\n')
````

## 1. 重新认识函数：函数也是对象
在 Python 中，一切皆对象——函数也不例外。函数属于 `function` 类型，可以被赋值、被当作参数传递、也可以作为返回值。

```python
# 函数也是对象
def welcome():
    print('你好啊！')

print(type(welcome))   # <class 'function'>

# 可以动态添加属性
welcome.author = 'Python学习小组'
welcome.version = 2.0
print(welcome.__dict__)

# 可以赋值给变量
say_hello = welcome
say_hello()            # 同样输出 '你好啊！'
```

```python
# 可变对象与不可变对象作为参数的区别
def modify(data):
    print(f'函数内接收: {data}, id: {id(data)}')
    data = 888        # 不可变对象：修改后指向新对象
    print(f'修改后: {data}, id: {id(data)}')

a = 666
print(f'调用前 a: {a}, id: {id(a)}')
modify(a)
print(f'调用后 a: {a}, id: {id(a)}')   # a 未变

print('---')
def modify_list(lst):
    print(f'接收列表: {lst}, id: {id(lst)}')
    lst.append(99)    # 可变对象：原地修改
    print(f'修改后: {lst}, id: {id(lst)}')

b = [10, 20, 30]
print(f'调用前 b: {b}, id: {id(b)}')
modify_list(b)
print(f'调用后 b: {b}, id: {id(b)}')   # b 被修改了！
```

```python
# 函数作为参数 和 作为返回值（高阶函数）
def greet():
    print('你好！')

def call_func(func):
    print('call_func 开始调用')
    func()

call_func(greet)

# 函数作为返回值
def outer():
    def inner(msg):
        print(msg)
    return inner

my_print = outer()
my_print('Python 函数进阶')
```

## 2. 多返回值与参数的打包/解包
Python 函数可以返回多个值，实际上是以元组形式打包；`*args` 和 `**kwargs` 则分别打包位置参数和关键字参数。调用时可用 `*` 和 `**` 解包元组/字典为独立参数。

```python
# 多返回值
def calculate(x, y):
    return x + y, x - y, x * y

res = calculate(10, 5)
print(res)           # (15, 5, 50)
a, b, c = calculate(10, 5)
print(a, b, c)       # 15 5 50
```

```python
# 打包接收参数
def show_info(*args, **kwargs):
    print('位置参数:', args)
    print('关键字参数:', kwargs)

# 解包传递参数
nums = (10, 20, 30)
person = {'name': '张三', 'age': 25}
show_info(*nums, **person)
```

## 3. 高阶函数与匿名函数 (lambda)
高阶函数指参数或返回值为函数的函数，它让代码更灵活。匿名函数用 `lambda` 定义，适合简单的单行表达式，常用于 map、filter 等场景。

```python
# 高阶函数：参数为函数
def log(func, text):
    print(func(text))

def info(msg):
    return '[提示] ' + msg
def error(msg):
    return '[错误] ' + msg

log(info, '文件保存成功')
log(error, '用户不存在')

# 匿名函数 lambda
def calculate(func, a, b):
    return func(a, b)

print(calculate(lambda x, y: x + y, 30, 10))  # 40
print(calculate(lambda x, y: x - y, 30, 10))  # 20

# 带有条件表达式的 lambda
is_adult = lambda age: '成年' if age >= 18 else '未成年'
print(is_adult(20), is_adult(15))
```

## 4. 数据处理四剑客：map / filter / sorted / reduce
这四个函数是函数式编程风格的代表，配合 lambda 可以简洁地对序列进行转换、过滤、排序和归并。

```python
# map：对每个元素执行相同操作
nums = [10, 20, 30, 40]
doubled = list(map(lambda x: x * 2, nums))
print('map x2:', doubled)

# filter：筛选符合条件的元素
nums = [10, 20, 30, 40, 50]
big = list(filter(lambda x: x > 30, nums))
print('filter >30:', big)

# filter 特殊用法：不传函数则自动过滤假值
data = [0, 1, '', 'hello', [], 5]
print('filter(None):', list(filter(None, data)))
```

```python
# sorted：灵活排序
names = ['python', 'java', 'go', 'rust']
print('按长度排序:', sorted(names, key=len))

persons = [
    {'name': '张三', 'age': 20},
    {'name': '李四', 'age': 18},
    {'name': '王五', 'age': 22}
]
print('按年龄排序:', sorted(persons, key=lambda p: p['age']))
print('最年长:', max(persons, key=lambda p: p['age']))
print('最年轻:', min(persons, key=lambda p: p['age']))
```

```python
# reduce：累积归并（需从 functools 导入）
from functools import reduce

nums = [1, 2, 3, 4, 5]
total = reduce(lambda a, b: a + b, nums, 10)  # 初始值10
print('累加结果:', total)

strs = ['ab', 'cd', 'ef']
concat = reduce(lambda a, b: a + b, strs)
print('字符串拼接:', concat)
```

## 5. 列表推导式（及字典、集合推导式）
列表推导式是 Python 的简洁语法，用于从可迭代对象快速生成列表，可包含条件筛选。字典和集合也有类似推导式，但元组推导式得到的是生成器。

```python
# 列表推导式
nums = [10, 20, 30, 40]
squared = [n**2 for n in nums]
print('平方:', squared)

# 带条件的列表推导式
even_squares = [n**2 for n in nums if n % 2 == 0]
print('偶数平方:', even_squares)

# 字典推导式
names = ['张三', '李四', '王五']
scores = [85, 92, 78]
score_dict = {names[i]: scores[i] for i in range(len(names))}
print('成绩字典:', score_dict)

# 集合推导式
unique = {n % 3 for n in range(10)}
print('集合推导式:', unique)

# 元组推导式？其实是生成器对象！
gen = (n**2 for n in nums)
print('生成器:', gen, list(gen))
```

## 7. 浅拷贝 vs 深拷贝
直接赋值不会复制对象，只是多个变量指向同一内存地址。对于可变对象，我们需要拷贝来避免相互干扰。
- **浅拷贝**（`copy.copy()`）：仅复制外层容器，内部元素仍指向原对象。
- **深拷贝**（`copy.deepcopy()`）：递归复制所有可变对象，完全独立。

```python
import copy

# 直接赋值：共享同一列表
a = [10, 20, [30, 40]]
b = a
b[2][0] = 99
print('直接赋值后 a:', a)   # 受到影响

# 浅拷贝：新外层容器，内层仍共享
a = [10, 20, [30, 40]]
c = copy.copy(a)
c[2][0] = 99
print('浅拷贝后 a:', a)     # 内层仍受影响

# 深拷贝：完全独立
a = [10, 20, [30, 40]]
d = copy.deepcopy(a)
d[2][0] = 99
print('深拷贝后 a:', a)     # 不受影响
```

## 8. 四种作用域与 LEGB 规则
Python 变量查找遵循 **L**ocal → **E**nclosing → **G**lobal → **B**uilt-in 的顺序。理解作用域是掌握闭包和装饰器的基础。

```python
# 局部作用域
def test():
    x = 10
    print(x)
test()
# print(x)  # 报错

# Enclosing（外层作用域）
def outer():
    y = 20
    def inner():
        print(y)   # 可读取外层变量
    inner()
outer()

# 修改外层变量需 nonlocal
def outer2():
    count = 0
    def inner():
        nonlocal count
        count += 1
        print(count)
    return inner
f = outer2()
f(); f()  # 1, 2

# 全局作用域
g = 100
def show_global():
    print(g)
show_global()

def modify_global():
    global g
    g = 200
modify_global()
print(g)  # 200
```

## 9. 闭包 (Closure)
闭包 = 内层函数 + 它所引用的外层变量。即使外层函数已经执行完毕，闭包仍能“记住”那些变量的状态。这是装饰器的核心基础。

```python
# 闭包经典例子：计数器
def make_counter():
    count = 0
    def counter():
        nonlocal count
        count += 1
        return count
    return counter

cnt1 = make_counter()
print(cnt1(), cnt1(), cnt1())  # 1 2 3
cnt2 = make_counter()          # 新的闭包，独立状态
print(cnt2())                  # 1

# 闭包如何存储变量？
print('闭包单元:', cnt1.__closure__)
print('存储的值:', cnt1.__closure__[0].cell_contents)
```

```python
# 闭包应用：配置化美化打印
def make_beauty(char, n):
    def show(msg):
        print(char * n + msg + char * n)
    return show

star_print = make_beauty('*', 3)
star_print('你好')          # ***你好***
at_print = make_beauty('@', 5)
at_print('Python')         # @@@@@Python@@@@@

# 等价于用类实现
class Beauty:
    def __init__(self, char, n):
        self.char = char
        self.n = n
    def show(self, msg):
        print(self.char * self.n + msg + self.char * self.n)

b = Beauty('#', 4)
b.show('闭包')
```

## 10. 装饰器（Decorator）
装饰器可以在不修改原函数代码的情况下增强其功能，是闭包和高级函数的经典应用。Python 支持用 `@` 语法糖优雅地使用装饰器。

### 10.1 函数装饰器基础

```python
# 定义一个简单的函数装饰器
def log_call(func):
    def wrapper(*args, **kwargs):
        print(f'[LOG] 调用 {func.__name__}，参数：{args}, {kwargs}')
        result = func(*args, **kwargs)
        print(f'[LOG] 结果：{result}')
        return result
    return wrapper

# 使用语法糖 @ 装饰函数
@log_call
def add(x, y):
    return x + y

print(add(3, 7))

# 等价于手动装饰：add = log_call(add)
```

```python
# 带参数的装饰器（三层嵌套）
def repeat(times):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(times):
                result = func(*args, **kwargs)
            return result
        return wrapper
    return decorator

@repeat(3)
def greet(name):
    print(f'你好，{name}！')

greet('小明')   # 打印三次

# 多个装饰叠加（由下往上装饰）
def bold(func):
    def wrapper():
        return '<b>' + func() + '</b>'
    return wrapper

def italic(func):
    def wrapper():
        return '<i>' + func() + '</i>'
    return wrapper

@bold
@italic
def get_text():
    return 'Python'

print(get_text())  # <b><i>Python</i></b>
```

### 10.2 类装饰器
实现了 `__call__` 方法的类的实例可以像函数一样被调用，因此也能作为装饰器。类装饰器在处理带参数场景时往往更简洁。

```python
# 基本的类装饰器
class Logger:
    def __call__(self, func):
        def wrapper(*args, **kwargs):
            print(f'调用 {func.__name__}')
            return func(*args, **kwargs)
        return wrapper

@Logger()         # 先实例化，再作为装饰器
def multiply(a, b):
    return a * b

print(multiply(6, 7))
```

```python
# 带参数的类装饰器
class Prefix:
    def __init__(self, prefix):
        self.prefix = prefix

    def __call__(self, func):
        def wrapper(*args, **kwargs):
            print(f'{self.prefix} 开始执行')
            result = func(*args, **kwargs)
            print(f'{self.prefix} 执行完毕')
            return result
        return wrapper

@Prefix('【模块A】')
def compute():
    print('计算中...')
    return 42

compute()
```

## 11. 类型注解 (Type Hints)
Python 3.5+ 支持为变量和函数添加类型注解，提升代码可读性和 IDE 智能提示。注解不会强制类型检查，但可配合 mypy 等工具进行静态分析。

```python
# 基本变量注解
name: str = '张三'
age: int = 30
price: float = 19.9
is_student: bool = False

# 容器注解（需 Python 3.9+ 或 from typing）
scores: list[int] = [85, 90, 78]
cities: set[str] = {'北京', '上海'}
person: dict[str, int] = {'张三': 25, '李四': 30}
# 混合类型
data: list[str | int] = ['apple', 3, 'banana']

# 元组注解特殊：可指定每个位置类型或任意个数
point: tuple[int, int] = (10, 20)
numbers: tuple[int, ...] = (1, 2, 3, 4)
```

```python
# 函数类型注解
def add(x: int, y: int) -> int:
    return x + y

def process(data: list[int]) -> tuple[int, int, float]:
    max_val = max(data)
    min_val = min(data)
    avg_val = sum(data) / len(data)
    return max_val, min_val, avg_val

# 查看注解
print(add.__annotations__)
print(process.__annotations__)

# 可变参数注解
def log_messages(*msgs: str) -> None:
    for msg in msgs:
        print(msg)

log_messages('开始', '处理中', '结束')
```
